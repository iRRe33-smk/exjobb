```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
def sigmoid_func(x, I, D):
    s = 1 / (1 + np.exp(-(x/I) + D))
    return s
```

```python
def sigmoid_func_paradigm(x, I, D):
    s = 1 / (1 + np.exp(-(x/I) + D)) +  1 / (1 + np.exp(-(x/I) + 3*D))
    return s
```

```python
x = np.linspace(-2,8)
I = 1
D = 3
plt.rcParams["figure.figsize"] = (15,5)
plt.plot(x, sigmoid_func(x, I, D))
```

```python
x = np.linspace(-2,15)
plt.rcParams["figure.figsize"] = (15,5)
plt.plot(x, sigmoid_func_paradigm(x, I, D))
```
