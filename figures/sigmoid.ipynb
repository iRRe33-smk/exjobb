```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
def sigmoid_func(x, I, D):
    s = 1 / (1 + np.exp(-(x/D) + I))
    return s
```

```python
x = np.linspace(0,4)
I = 6.5
D = 0.3
plt.rcParams["figure.figsize"] = (15,5)
plt.plot(x, sigmoid_func(x, I, D))
```
